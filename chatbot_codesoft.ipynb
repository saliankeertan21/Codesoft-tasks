```python
responses={
    "greeting": "Hey there Welcome I am your assistant, How can I help you?",
    "name":"I am your python help bot",
    "help":"You can ask me basic questions",
    "bye": " it was nice meeting you,Goodbye",
    "default": " Sorry i didnt understand that"
}

while True:
    user_input= input("You: ").lower()

    if "hi" in user_input or "hello" in user_input:
        print("Bot:  ",responses["greeting"])

    elif "name" in user_input:
        print("Bot: ", responses['name'])

    elif "help" in user_input:
        print("Bot:  ",responses['help'])

    elif "bye" in user_input:
        print("Bot: ",responses['bye'])
        break

    else:
        print("Bot: ",responses['default'])
```
